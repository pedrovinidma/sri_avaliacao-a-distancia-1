```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
dataset = load_dataset("celsowm/cnn_news_ptbr")

print(dataset)
```

```python
cont = 1
for i, x in enumerate(dataset["train"]):
      if cont <= 30:
        print("\n", x)
        cont = cont + 1
```

```python
import pandas as pd

# Lista para receber os dados selecionados
dados = []

# Percorre o dataframe
for i, x in enumerate(dataset["train"]):
      # Seleciona somente 30 documentos
      if i < 30:
        # Adiciona o id e o documento
        dados.append({
            "id": i + 1,
            "documento": x["resumo"]
        })

# Cria o DataFrame para poder salvar os dados
df = pd.DataFrame(dados)

# Salva como CSV
df.to_csv("documentos.csv", index=False, encoding="utf-8", sep=';')
```
